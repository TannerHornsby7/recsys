```python
# create "get movie" function which takes in the movie lens id and uri for tmdb
import requests
import os

os.getenv('TMDB_BEARER_TOKEN')
```

```python

def get_movie_details(movie_lens_id):
    tmdb_id = links_df[links_df['movieId'] == movie_lens_id]['tmdbId'].values[0]
    url = f"https://api.themoviedb.org/3/movie/{tmdb_id}?language=en-US"
    headers = {
        "accept": "application/json",
        "Authorization": f"Bearer {os.getenv('TMDB_BEARER_TOKEN')}"
    }
    response = requests.get(url, headers=headers)
    return response.json()

# get the movie details for a random movie
movie_lens_id = df.sample(1)['ITEM_ID'].values[0]
movie_details = get_movie_details(movie_lens_id)
movie_details, get_movie_details(1608)

# append the path resources to https://image.tmdb.org/t/p/original/bzxiGlVjlF7sDNviWOdFl1w10cP.jpg
```
